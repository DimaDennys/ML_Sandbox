```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

## Froth Flotation Operating Data

```python

data = np.array([
    [120, 800, 35, 60, 8.5],
    [130, 820, 36, 62, 8.6],
    [125, 810, 34, 59, 8.4],
    [200, 1000, 45, 80, 10.0],
    [210, 1020, 46, 82, 10.2],
    [195, 980, 44, 78, 9.8],
    [90, 700, 30, 50, 7.2],
    [95, 720, 31, 52, 7.4],
    [85, 690, 29, 48, 7.1]
])

labels = [
    'Normal-A','Normal-B','Normal-C',
    'High-A','High-B','High-C',
    'Low-A','Low-B','Low-C'
]
```

## Hierarchical Clustering and Dendrogram

```python

Z = linkage(data, method='ward')

plt.figure(figsize=(10,6))
dendrogram(Z, labels=labels, leaf_rotation=45)
plt.title('Froth Flotation Operating Regimes')
plt.xlabel('Operating Conditions')
plt.ylabel('Distance')
plt.show()
```
