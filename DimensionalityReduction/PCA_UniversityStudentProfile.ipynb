For the University project, let's shift from "sensor data" to "Student Academic Profiles." In this synthetic dataset, we create hidden correlations: for example, students who are good at Math are usually good at Physics and CS, while students who are good at Literature are often strong in History. PCA will help us "compress" these subjects into two main dimensions: STEM vs. Humanities.

1. Generating the Synthetic Student Data
We will simulate 500 students with grades (0–100) across 6 subjects.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 1. Create Synthetic Student Data
np.random.seed(7)
n_students = 500

# Base "Intelligence" factors (Hidden variables)
stem_aptitude = np.random.normal(70, 10, n_students)
humanities_aptitude = np.random.normal(65, 12, n_students)

# Generate grades based on these aptitudes + some noise
data = {
    'Math': stem_aptitude + np.random.normal(0, 5, n_students),
    'Physics': stem_aptitude * 0.9 + np.random.normal(0, 5, n_students),
    'Comp_Sci': stem_aptitude * 1.1 + np.random.normal(0, 7, n_students),
    'Literature': humanities_aptitude + np.random.normal(0, 5, n_students),
    'History': humanities_aptitude * 0.8 + np.random.normal(0, 6, n_students),
    'Art': humanities_aptitude * 0.7 + np.random.normal(0, 10, n_students)
}

df_students = pd.DataFrame(data).clip(0, 100) # Ensure grades stay between 0-100

print(df_students.head(5))
```

2. PCA Analysis Pipeline
In your Jupyter notebook, this section demonstrates how we collapse 6 subjects into 2 "Study Profiles."

```python
# 2. Standardize (Crucial so Math and Art are weighted equally)
scaler = StandardScaler()
scaled_grades = scaler.fit_transform(df_students)

# 3. Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_grades)

# Create a DF for the new "Profile" axes
df_pca = pd.DataFrame(data = principal_components, columns = ['PC1', 'PC2'])
```

3. Visualizing the "Student Map" (The Biplot)
A Biplot is the most important visual for University admin. It shows where students sit and which subjects "pull" them in certain directions.

```python
# Calculate Loadings (how much each subject contributes to PC1 and PC2)
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
subjects = df_students.columns

plt.figure(figsize=(10, 8))

# Plot the students as dots
plt.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.3, c='gray')

# Plot the Subject Vectors (Loadings)
for i, subject in enumerate(subjects):
    plt.arrow(0, 0, loadings[i, 0]*3, loadings[i, 1]*3, color='r', alpha=0.8, head_width=0.1)
    plt.text(loadings[i, 0]*3.5, loadings[i, 1]*3.5, subject, color='black', ha='center', va='center', fontweight='bold')

plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} Variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} Variance)')
plt.title('University Student Academic Profiling (PCA Map)')
plt.grid(True)
plt.show()
```
