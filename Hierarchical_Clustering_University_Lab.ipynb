```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage


```

## Reaction Experiment Data

```python

data = np.array([
    [80, 5, 60, 0.5, 72],
    [82, 5.2, 62, 0.55, 74],
    [78, 4.8, 58, 0.48, 71],
    [120, 10, 30, 1.0, 88],
    [122, 10.2, 32, 1.1, 90],
    [118, 9.8, 28, 0.95, 87],
    [60, 3, 90, 0.3, 60],
    [62, 3.2, 88, 0.32, 62]
])

labels = [
    'Moderate-1','Moderate-2','Moderate-3',
    'High-1','High-2','High-3',
    'Low-1','Low-2'
]
```

## Hierarchical Clustering and Dendrogram

```python

Z = linkage(data, method='ward')

plt.figure(figsize=(10,6))
dendrogram(Z, labels=labels, leaf_rotation=45)
plt.title('University Lab Reaction Clusters')
plt.xlabel('Experiments')
plt.ylabel('Distance')
plt.show()
```
